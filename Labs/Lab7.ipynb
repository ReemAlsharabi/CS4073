```python
import pandas as pd

# Read the data
data = pd.read_csv('survey (1).csv')
data.head()
```

### Data cleaning

```python
# Replace NaN values with the mean of the column
data = data.fillna(data.mean())
# Check if any remaining NaN
nulls = data.isnull().sum()
print(nulls)
```

```python
data.head()
```

### Likert chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the mean of each Likert scale column
likert_means = data.mean()

# Define the Likert scale labels
likert_labels = ['Strongly Disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly Agree']

# Create a bar plot
plt.bar(range(len(likert_means)), likert_means)

# Add the Likert scale labels to the x-axis
plt.xticks(range(len(likert_means)), data.columns)

# Add labels and title
plt.xlabel('Likert Scale Questions')
plt.ylabel('Mean Rating')
plt.title('Likert Chart')

# Show the plot
plt.show()
```

It seems that a majority of the respondents have a positive sentiment or agreement towards the problem, while a smaller portion may hold a more neutral stance.
